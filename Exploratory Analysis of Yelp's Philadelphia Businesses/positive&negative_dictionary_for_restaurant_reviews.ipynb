```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, f1_score, precision_recall_fscore_support
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

le = LabelEncoder()
df['sentiment_label'] = le.fit_transform(df['sentiment'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['sentiment_label'], test_size=0.2, random_state=42)

# Vectorize the text data using tf-idf
vectorizer = TfidfVectorizer(stop_words='english')
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Fit a logistic regression model to the training data
logreg = LogisticRegression(random_state=0, solver='lbfgs', multi_class='auto', n_jobs=-1)
logreg.fit(X_train_vectorized, y_train)

# Make predictions on the test data
logreg_pred = logreg.predict(X_test_vectorized)

# Calculate evaluation metrics
logreg_precision, logreg_recall, logreg_fscore, logreg_support = precision_recall_fscore_support(y_test, logreg_pred, average='macro')
logreg_acc = logreg.score(X_test_vectorized, y_test)
logreg_f1score_score_last3 = f1_score(y_test, logreg_pred, average='macro', labels=[0, 1, 2])

# Print the evaluation metrics
print('Accuracy:', logreg_acc)
print('Precision:', logreg_precision)
print('Recall:', logreg_recall)
print('F1 Score:', logreg_fscore)

# Print the confusion matrix
cm_logreg = confusion_matrix(y_test, logreg_pred)
print('Confusion matrix:')
print(cm_logreg)

# Get the coefficients of the logistic regression model
feature_names = vectorizer.get_feature_names()
coefficients = logreg.coef_[0]
coef_df = pd.DataFrame({'feature': feature_names, 'coefficient': coefficients})
coef_df.sort_values(by='coefficient', ascending=False, inplace=True)
print("Words with highest positive weights:")
print(coef_df.head(20))
print("Words with lowest negative weights:")
print(coef_df.tail(20))
```

```python
positive_dictionary = []
negative_dictionary = []

for index, row in coef_df.iterrows():
    if row['coefficient'] > 1.:
        positive_dictionary.append(row['feature'])
    if row['coefficient'] < -2:
        negative_dictionary.append(row['feature'])
```

```python
def classify_review(text):
    doc = nlp(text)
    # get noun chunks and adjectival phrases
    phrases = []
    for chunk in doc.noun_chunks:
        phrases.append(chunk.text)
    for token in doc:
        if token.pos_ == 'ADJ':
            phrases.append(token.text)

    # count the number of positive and negative words in the text
    num_positive = 0
    num_negative = 0
    for phrase in phrases:
        if phrase in positive_dict:
            num_positive += positive_dict[phrase]
        if phrase in negative_dict:
            num_negative += negative_dict[phrase]

    # classify the review based on the number of positive and negative words
    if num_positive > num_negative:
        return 'Positive'
    elif num_positive < num_negative:
        return 'Negative'
    else:
        return None
```
